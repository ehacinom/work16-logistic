```python
users_init = ('INSERT INTO users (UserID, IP, events, elapsed, period, cv)'
              'SELECT UserID, group_concat(IP), count(*), (max(t)-min(t)), '
              '(max(t)-min(t))/count(t), sum(ConversionTagID), '
              'FROM logs GROUP BY logs.UserID')
with sqlite3.connect() as con:
    cur = con.cursor()
    cur.execute(users_init)
```

```python
profile_init = ('INSERT INTO profile (UserID, IP, events, elapsed, period, cv, times)'
                'SELECT UserID, group_concat(IP), count(*), (max(t)-min(t)), '
                '(max(t)-min(t))/count(t), sum(ConversionTagID), group_concat(t) '
                'FROM logs GROUP BY logs.UserID')
with sqlite3.connect() as con:
    cur = con.cursor()
    cur.execute(profile_init)
```

Count.

```python
def countandplot(ee, kind='count and plot elapsed, events', n_events = 300):
    print kind

    eex, eey = zip(*ee)
    h, x, y = np.histogram2d(eex, eey, bins=[n_events*10,n_events])
    xi, xf, yi, yf = x[0]/secinday, x[-1]/secinday, y[0], y[-1]

    fig, ax = plt.subplots(figsize=(13,10))
    # transpose the array at h
    im = ax.imshow(h.T, origin = "lower", norm=matplotlib.colors.LogNorm(), interpolation='none', extent=[xi, xf, yi, yf], cmap=plt.cm.jet) 
    ax.set_title(r'$log(count := f(\Delta t, N)$', fontsize=16)
    ax.set_xlabel('elapsed')
    ax.set_ylabel('events')
    ax.set_aspect((xf-xi)/(yf-yi)) # square

    fig.colorbar(im)
    plt.show()
```
