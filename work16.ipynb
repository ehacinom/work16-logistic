###Prelim definition
The place where preliminary classification is flagged is **prelim**, in the profile table. This is mildy a man-made construct, and partly definitive classification. Classification, if a UserID falls under several of the preconditions, will be sent to the larger, absolute value. For instance, if a User can be marked both 5 (human) and -1 (highf, or high frequency), it will be marked 5/human.

The current definitions we have are (name, prelim, condition for classification):
- human, 5, cv = 2 (3572)
- phone, 2, MobileDevice > 1 (48716)
- highf, -1, f >= 1 (48613)
- highn, -2, n >= 300 (1234)
- robot, -5, IP \in crawlers (148)

```python
def regexp(expr, item):
    result = re.match(expr, item)
    if result is not None: 
        return result.group(1)
    else: return result

# test
print regexp('^https?:\/\/(.*?)\/', 'http://www.orixlife.jp/product/ajax/')

with sqlite3.connect(dbname) as con:
    con.create_function("REGEXP", 2, regexp)
    cur = con.cursor()
    cmd = "UPDATE logs SET Referrer = Referrer REGEXP '^https?:\/\/(.*?)\/' WHERE Referrer is not NULL"
    cur.execute(cmd)
```

```python
def countandplot(cols=[], 
                 bins = defaultdict(lambda: 10), 
                 ranges = defaultdict(lambda: None), 
                 normed = defaultdict(lambda: False),
                 plotnorm = matplotlib.colors.LogNorm()):
    '''
    cols:   list of string column names
            limited to COLS variable, defined in previous cell
    bins:   {array: <int>}
    ranges: {array: [[xmin, xmax], [ymin, ymax]]}
    normed: {array: True}
    plotnorm: 
    '''
    # check
    n = len(cols)
    if not n: 
        print 'ERROR: No columns.'
        return
    if not set(cols).issubset(COLS):
        print 'ERROR: Check columns are in your dataframe.'
        return
    
    # set up six populations
    robot, highn, highf, phone, human, other = (pd.DataFrame() for i in xrange(6))
    for i in xrange(n):
        robot[cols[i]] = pre[pre['prelim'] == -5][cols[i]]
        highn[cols[i]] = pre[pre['prelim'] == -2][cols[i]]
        highf[cols[i]] = pre[pre['prelim'] == -1][cols[i]]
        phone[cols[i]] = pre[pre['prelim'] == 2][cols[i]]
        human[cols[i]] = pre[pre['prelim'] == 5][cols[i]]
        other[cols[i]] = prelim[cols[i]]

    # helpers
    def plotter(arr, name):
        # no nulls (np.histogram can't deal) and won't accept the dataframe with wrong index?
        arr.dropna(inplace=True)
        arr = arr.as_matrix()

        # histogram
        H, edges = np.histogramdd(arr, bins=bins[name], range=ranges[name], normed=normed[name])   

        # edges
        xi, xf, xx = edges[0][0], edges[0][-1], edges[0][1] - edges[0][0]
        yi, yf, yy = edges[1][0], edges[1][-1], edges[1][1] - edges[1][0]

        # plot
        ax = axes.flat[sp[name]]
        im = ax.imshow(H.T, origin = 'lower', norm=plotnorm, extent=[xi, xf, yi, yf], cmap=plt.cm.jet)
        ax.set_title(name, fontsize=11.5)
        #ax.set_xlabel(cols[0])
        #ax.set_ylabel(cols[1])
        ax.set_aspect((xf-xi)/(yf-yi)) # square
        
        return im

    # plot 
    fig, axes = plt.subplots(2, 3, squeeze=True)
    sp = {'robot': 0, 'highn': 1, 'highf': 2, 'phone': 3, 'human': 4, 'other': 5} # subplots
    plotter(robot, 'robot')
    plotter(highn, 'highn')
    plotter(highf, 'highf')
    plotter(phone, 'phone')
    plotter(human, 'human')
    im = plotter(other, 'other') # for the colorbar
    fig.subplots_adjust(right = 0.8)
    colorbarax = fig.add_axes([0.85, 0.15, 0.03, 0.75])
    fig.colorbar(im, cax=colorbarax)
    fig.set_size_inches(18,9)
    fig.suptitle('Preliminary classification population characteristics', x=0.5, y=0.98, fontsize=16)
    fig.text( x=0.5, y=0.02, s=cols[0], fontsize=14.5)
    fig.text( x=0.02, y=0.5, s=cols[1], fontsize=14.5)
    plt.show()
```
